```python
import numpy as np
import matplotlib.pyplot as plt
import os

import pyximport; pyximport.install()
from smcTranslocatorCircular_AsymmetricDissociatesOrBypasses import smcTranslocatorCircular_DissociatesOrBypasses

import os 
import shutil

import contact_map_generator_from_shortest_path as cmgsp
from contact_map_generator_from_shortest_path import zoomArray

import sys
sys.setrecursionlimit(4000)

from multiprocessing import Pool
import pickle

from matplotlib.gridspec import GridSpec
from itertools import product
from scipy.ndimage.filters import gaussian_filter
import matplotlib.colors as clr 
```

## 1D simulation setup function

```python

def initModel(knockOffProb = 1,kOriToTer_kTerToOri_kBypass=[0.01,0.01,0.04],parSsites=[-1,-4,-5,-6,-27,4,17],\
              wind=0.33,k_slow=0,terSites=[2000,2050],stallList = [],parSstrengths=None,\
             SEPARATION=200,LIFETIME = 2000,N = 4000,BASE_STOCHASTICITY=0.1):

    """
    This function initializes the loop extrusion factor simulation and generates and SMC 
    translocation object. 
    
    Parameters:
    -----------
    knockOffProb : float
        Probability that a LEF (loop extruding factor) after stalling is dissociated
        
    kOriToTer_kTerToOri_kBypass : 3-tuple of floats
        Probability that upon a "collision" event between two or more LEFs that a LEF will
        stall (i.e. dissociate if the knockOffProb=1), or bypass one another. 
        The first term is the unloading rate for subunits moving in the Ori to ter direction. 
        The second term is for subunits moving in the ter to ori direction.
        The last term is the probability that the subunit will step onto an occupied site.
        
    parSsites : list of floats
        Determines the angular position of the parS sites
        
    wind : float
        Value between 0 and 1 determining the bias in speeds towards origin or terminus. 
        A value of 1 means that it's effectively a 1-sided LEF and a value of 0 is no bias.
        
    k_slow : can ignore this
    
    terSites : list of int
        Determines the range of unloading sites on the genome
        
    stallList : list of int
        Determines positions of places where extrusion is stalled (can ignore)
        
    parSstrengths : list of floats
        List has to be the same size as `parSsites` list. Determines the relative strengths
        of condensin loading to each site.
        
    SEPARATION : float
        Average separation between LEFs (in monomer units). It is genome length/ number of LEFs.
    
    LIFETIME : float
        The average numer of simulation time-steps before a LEF dissociates.
        
    N : int
        Number of LEFs
        
    BASE_STOCHASTICITY : float
        Value between 0 and 1 determining the probability per simulation time step of a LEF moving
        lattice sites
    """
    
    birthArray = np.ones(N, dtype=np.double)  
    deathArray = np.zeros(N, dtype=np.double) + 1. / (LIFETIME/BASE_STOCHASTICITY)
    stallDeathArray = np.zeros(N, dtype=np.double) + knockOffProb
    stallLeftArray = np.zeros(N, dtype = np.double)
    stallRightArray = np.zeros(N, dtype = np.double)
    stepArrayR = BASE_STOCHASTICITY*np.ones(N, dtype=np.double) 
    stepArrayL = BASE_STOCHASTICITY*np.ones(N, dtype=np.double)
    
    deathList = [x for x in np.arange(terSites[0],terSites[1])]
    for i in deathList:
        deathArray[i] = 0.05*BASE_STOCHASTICITY

    for i in stallList:
        stallLeftArray[i] = 0.8
        stallRightArray[i] = 0.8
        
    # set parS locations, and compute the slowing down values
    if parSstrengths is None:
        parSstrengths = [16000/len(parSsites)]*len(parSsites)
    parS_arrayR = []
    parS_arrayL = []
    for s,ss in zip(parSsites,parSstrengths):
        if np.mod(int(s/360*N),N) < terSites[0]:
            parS_arrayR.append(np.mod(int(s/360*N),N))
        else:
            parS_arrayL.append(np.mod(int(s/360*N),N))            
        birthArray[np.mod(int(s/360*N),N)] = ss
    
    
    # set baseline translocation asymmetry
    stepArrayL[0:terSites[0]] *= (1-wind)
    stepArrayR[terSites[1]:N] *= (1-wind) 
    
    # set stages of asymmetry
    parS_arrayR = np.sort(parS_arrayR)
    parS_arrayL = np.sort(parS_arrayL)[::-1]
    
    # ALL SITES IN BETWEEN parS GET SLOWDOWN METHOD (depracated)
    if len(parS_arrayR)>1: 
        for l, st,nd in zip(np.arange(1,len(parS_arrayR)),parS_arrayR[0:-1],parS_arrayR[1:]):
            stepArrayR[st:nd] *= (1-k_slow)
            stepArrayL[st:nd] *= (1-k_slow)         
    if len(parS_arrayL)>1: 
        for l, st,nd in zip(np.arange(1,len(parS_arrayL)),parS_arrayL[1:],parS_arrayL[0:-1]):
            stepArrayR[st:nd] *= (1-k_slow)
            stepArrayL[st:nd] *= (1-k_slow)
            
    smcNum = N // SEPARATION
    pauseArrayR = 1-stepArrayR
    pauseArrayL = 1-stepArrayL
    SMCTran = smcTranslocatorCircular_DissociatesOrBypasses(birthArray, deathArray, stallLeftArray, \
                                                    stallRightArray, pauseArrayL, pauseArrayR, \
                                                    stallDeathArray, kOriToTer_kTerToOri_kBypass,  smcNum)
    
    return SMCTran

SMCTran = initModel() 
```

## Setting up the extrusion parameters, and strain parameters

```python
# sweep system values
samps_per_core = 400  # number of samples per CPU 
initialization_steps = 100000 # number of equilibration steps
steps_per_sample = 1000 # steps taken for each CPU (i.e. independent simulation)
num_cores = 30 # total number of cores used

# parameters for sweep
N = 4040 # genome size in monomers
BASE_STOCHASTICITY = 0.05 # probability of an extruder taking a step (per simulation time-step)
k_knock_list = [0.0025] # probability of an extruder that is "blocked" by another extruder from dissociating from the genome
k_bypass_list = [0.05] # probability of an extruder that is "blocked" to bypass that extruder
base_stochasticity = [BASE_STOCHASTICITY] # same as above (in case you want to sweep the stepping probability)
life_times = [N/2] # dissociation rate of the extruder 
separation_list = [101] # average separation of extruders (i.e. this sets the total number of condensins)
parS_sites_list = [[-59,-27],[-59]] # locations of parS sites in genome angular coordinates
parS_strength_list = [8000] # relative strength of the parS sites to regular non-parS monomers (this value is divided by the number of sites)
wind_val_list = [0.35] # the direction bias in the speeds (i.e. sets the "tilt" of the secondary diagonals)
ter_sites = [[1950,2050]] # putative location of dissociation sites (best guess)
comb_list = list(product(k_knock_list,k_bypass_list,life_times, 
             separation_list,parS_strength_list,parS_sites_list, wind_val_list,base_stochasticity,ter_sites))

len(comb_list), comb_list[0]
```

## Run simulations for each parameter combination

```python

for k_knock, k_bypass, lifetime, separation, parSval, parSsites, wind, BASE_STOCHASTICITY, ter_sites in comb_list:
    
    # make folders and file names
    save_folder = ('Bypass_ChIP_HiC_sims_'+
                    "_".join(['{}'.format(str(s) if s>0 else ('m'+str(-s))  )
                               for s in sorted(parSsites) ]))
    fname = "kd{}_kb_{}_sep_{}_parSstrength_{}_lifetime_{}_wind_{}_stoch_{}".format(np.round(k_knock,5),\
                                            np.round(k_bypass,2),int(separation),parSval,\
                                            int(lifetime),wind,BASE_STOCHASTICITY)
    
    tmp_folder = os.path.join(save_folder,fname)
    if os.path.exists(tmp_folder):
        continue
    if not os.path.exists(save_folder):
         os.mkdir(save_folder)
    if os.path.exists(tmp_folder):
        shutil.rmtree(tmp_folder)
    if not os.path.exists(tmp_folder):
        os.mkdir(tmp_folder)


    # generate giant list of SMC translocator objects       
    SMCTran_list = [initModel(knockOffProb = 1,kOriToTer_kTerToOri_kBypass=[k_knock,k_knock,k_bypass],parSsites=parSsites,\
          wind=wind,k_slow=0,terSites=ter_sites,stallList = [],parSstrengths=[parSval/len(parSsites)]*len(parSsites),\
         SEPARATION=separation,LIFETIME = int(lifetime/BASE_STOCHASTICITY),N = N, BASE_STOCHASTICITY=BASE_STOCHASTICITY)\
                    for _ in range(num_cores)]

    def doOne(idx): 

        SMCTran = SMCTran_list[idx]
        SMCTran.steps(initialization_steps)

        lef_list = []
        for s in range(samps_per_core):

            SMCTran.steps(steps_per_sample)
            x,y = SMCTran.getSMCs()
            lef_list.append([(x[j],y[j]) for j in range(len(x))])         

        pickle.dump(lef_list,open(os.path.join(tmp_folder,'loops_{}.pkl'.format(idx)),'wb'))
        
        # generate heat maps
        hmap = cmgsp.heatmap(np.zeros((404,404)),np.zeros((404,404)))
        for smcs_pairs in lef_list:
            #HiC, HiC_count = cmgsp.contact_map_generator_from_SMC_list(smcs_pairs)
            HiC, HiC_count = cmgsp.contact_map_generator_from_SMC_list_slow(smcs_pairs,N=4040)
            hmap.HiC += zoomArray(HiC,(404,404))
            hmap.HiC_count += 1
        try:       
            with open(os.path.join(tmp_folder,'hmap_{}.pkl'.format(idx)),'wb') as FULLNAME:
                pickle.dump(hmap,FULLNAME)
        except:
             print("Failed at {}".format(idx))
        return 1

    pool = Pool(num_cores)
    ret_val = pool.map(doOne, np.arange(num_cores))

    # check that all the returned values are "good"
    if not np.alltrue(ret_val):
        print('error')

    # plot results
    fig = plt.figure(figsize=(5*1.5,10))
    gs = GridSpec(2, 2, height_ratios=[5,20], width_ratios=[18,1], wspace=0.1,hspace=0.3)

    # get the SMC ChIP-seq profile
    # plot experiment 
    ax2 = plt.subplot(gs[0])
    #plt.ylim([1e3,8e4])
    
    # plot simulation 
    files = [x for x in os.listdir(tmp_folder) if 'loops' in x]
    SMCs = np.zeros(N)
    for f in files:
        block = pickle.load(open(os.path.join(tmp_folder,f),'rb'))
        SMCs += np.bincount(np.ndarray.flatten( np.array(block)),minlength=4040)

    plt.xlim([-2020,2020])
    plt.plot(np.arange(-2020,2020,4040/len(SMCs)),np.fft.fftshift(SMCs))
    plt.title('kd={} kb={} proc={} sep={} wind={}'.format(np.round(k_knock,5),k_bypass,int(lifetime),separation,wind))
    plt.ylabel('SMC enrichment (arb)')
#     plt.ylim([0,50000])
    
    
    # get the Hi-C profile
    files = [x for x in os.listdir(tmp_folder) if 'hmap' in x]
    for idx, f in enumerate(files):
        with open(os.path.join(tmp_folder,f),'rb') as FULLNAME:
            hmap = pickle.load(FULLNAME)
            if idx == 0:
                M = hmap.HiC
                C = hmap.HiC_count
            else:
                M+= hmap.HiC
                C += hmap.HiC_count
    ZZ = hmap.HiC/hmap.HiC_count
    ZZ = (np.sqrt(ZZ))
    shiftBy = -2000
    s = -shiftBy
    ss = -shiftBy//10
    plt.subplot(gs[2])
    plt.imshow( np.roll(np.roll(np.flipud(ZZ),-ss,axis=0),ss,axis=1),
               vmin=0,vmax=0.01,cmap='YlOrBr',extent=[-2020,2020,-2020,2020],aspect='equal')
    plt.xlabel("Genome position (kb)")
    plt.ylabel("Genome position (kb)")
    plt.xlim([-2020,2020])
    plt.ylim([-2020,2020])

    # color bar in it's own axis
    colorAx = plt.subplot(gs[3])
    norm = clr.Normalize(vmin=0, vmax=0.01)
    cb = plt.colorbar(plt.cm.ScalarMappable(norm=norm,cmap='YlOrBr'),cax = colorAx)
    cb.set_label('Contact frequency')
    plt.draw()
    plt.savefig(os.path.join(tmp_folder,'figure_'+fname+'.pdf'))
    plt.show()
```
